## Load RGB images, convert to HSV

```python
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline

inpath = PATH + 'forest_jpg/'

for fname in glob.glob(inpath + '*rgb.jpg')[0:1]:
    #print(fname)
    img = cv2.imread(fname)
    
    if img is None:
        print('Failed to load image file: ', fname)
        #sys.exit(1)
        pass
    
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        
        # plot HSV
        plt.axis('off')
        plt.subplot(1,3,1)
        plt.imshow(img[:,:,0], cmap='gray')
        plt.subplot(1,3,2)
        plt.imshow(img[:,:,1], cmap='gray')
        plt.subplot(1,3,3)
        plt.imshow(img[:,:,2], cmap='gray')
        print(img.shape)
        
```

## Plot Histogram

```python
# plot histograme
for d in range(3):
    plt.hist(img[:,:,d].flatten(), bins=256)
    plt.show()
```
